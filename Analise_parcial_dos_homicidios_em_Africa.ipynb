A analise de dados efectuada neste projecto, esta relacionado com um dataset que descreve o numero de homicidios cometido no intervalo de ano de 2006 a 2021 por 195 paises de 5 continentes. Nossa análise focou-se em verificar como o continente Africano estava acerca deste tema. Buscamos verificar os dados do continente Africano por subregiões e anala-los. Uma chamda de atenção neste dataset dispensado pelo professor (Ruhid), é que os paises não encaminharam os dados de forma sequencial.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Analisando dados da África Central - Middle Africa
```

```python
dados_ac = dados[dados['Subregion'] == 'Middle Africa']
dados_ac
```

```python
taxa_total = dados_ac['Rate'].sum()
max_taxa = dados_ac['Rate'].max()
min_taxa = dados_ac['Rate'].iloc[1]
media_taxa_total = dados_ac['Rate'].mean()
casos_total = dados_ac['Count'].sum()
max_casos = dados_ac['Count'].max()
min_casos = dados_ac['Count'].min()
media_casos_total = dados_ac['Count'].mean()
```

```python
print("Dados da Africa Central:")
print("O total de casos de suicído foi:", casos_total)
print("A taxa total foi:", taxa_total)
print("A média de casos foi de:", media_casos_total)
print("O pais com mais caso de homicídio foi Angola com", max_casos)
print("O pais com menos caso de homicídio foi São Tomé and Príncipe com ", min_casos)
print("A o pais com maior taxa foi da Central African Republic com", max_taxa)
print("A menor taxa foi dos Cameroon ", min_taxa)
print("A média da taxa do homicídio foi de", media_taxa_total)
```

```python
sns.barplot(x = 'Location', y = 'Rate', data = dados_ac, palette='tab10');
plt.xlabel('Location')
plt.ylabel('Taxa')
plt.title('Total da Taxa de homicídio')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x = 'Location', y = 'Count', data = dados_ac, palette='tab10');
plt.xlabel('Location')
plt.ylabel('Taxa')
plt.title('Total de casos de homicídio')
plt.xticks(rotation=90)
plt.show()
```
